```python
with open('../files/bert_tsne_3d.pkl', 'rb') as f:
    bert_tsne_3d = pkl.load(f)

with open(f'../files/sent2vec_tsne_3d.pkl', 'rb') as f:
    sent2vec_tsne_3d = pkl.load(f)
```

```python
number_of_rows = 2000
# Create subplots for 3D visualization
fig = plt.figure(figsize=(15, 7))

# BERT Embeddings in 3D
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.scatter(bert_tsne_3d[:number_of_rows, 0], bert_tsne_3d[:number_of_rows, 1], bert_tsne_3d[:number_of_rows, 2], color='blue')
ax1.set_title('BERT Embeddings')
ax1.grid(True)
 
# Sent2Vec Embeddings in 3D
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.scatter(sent2vec_tsne_3d[:number_of_rows, 0], sent2vec_tsne_3d[:number_of_rows, 1], sent2vec_tsne_3d[:number_of_rows, 2], color='red')
ax2.set_title('Sent2Vec Embeddings')
ax2.grid(True)
 
# Show plot
plt.show()
```

```python
with open('../files/bert_tsne_3d.pkl', 'wb') as f:
    pkl.dump(bert_tsne_3d, f)

with open('../files/sent2vec_tsne_3d.pkl', 'wb') as f:
    pkl.dump(sent2vec_tsne_3d, f)
```
